## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.DataFrame(pd.read_csv(mouse_metadata_path))
study_results = pd.DataFrame(pd.read_csv(study_results_path))

# Combine the data into a single dataset
study_meta_merge = pd.merge(study_results, mouse_metadata, on = 'Mouse ID')
# Display the data table for preview
study_meta_merge

```

```python
# Checking the number of mice.
study_meta_merge['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
study_meta_merge.sort_values(['Mouse ID', 'Timepoint'], axis = 'index', inplace = True)
smm_dupe = study_meta_merge[study_meta_merge.duplicated(['Mouse ID', 'Timepoint'])]


```

```python
# Optional: Get all the data for the duplicate mouse ID. 
smm_dupe
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
smm_nodupe = study_meta_merge.drop_duplicates(['Mouse ID', 'Timepoint'])
smm_nodupe
```

```python
# Checking the number of mice in the clean DataFrame.
smm_nodupe['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_groups = smm_nodupe.groupby(['Drug Regimen'])
summary_tumor_vol = pd.DataFrame(drug_groups['Tumor Volume (mm3)'].mean())
summary_tumor_vol.rename(columns = {'Tumor Volume (mm3)':'Mean'}, inplace = True)
summary_tumor_vol['Median'] = drug_groups['Tumor Volume (mm3)'].median()
summary_tumor_vol['Variance'] = drug_groups['Tumor Volume (mm3)'].var()
summary_tumor_vol['Std Dev'] = drug_groups['Tumor Volume (mm3)'].std()
summary_tumor_vol['SEM'] = drug_groups['Tumor Volume (mm3)'].sem()

summary_tumor_vol
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_tumor_vol2 = drug_groups['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])
summary_tumor_vol2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
select_ca = smm_nodupe.loc[smm_nodupe['Drug Regimen'] == 'Capomulin'] 
select_ra = smm_nodupe.loc[smm_nodupe['Drug Regimen'] == 'Ramicane']
select_in = smm_nodupe.loc[smm_nodupe['Drug Regimen'] == 'Infubinol']
select_ce = smm_nodupe.loc[smm_nodupe['Drug Regimen'] == 'Ceftamin']

select_drug = pd.concat([select_ca, select_ra, select_in, select_ce])
#select_drug

# Start by getting the last (greatest) timepoint for each mouse
last_vol_group = select_drug.groupby(['Mouse ID'])
last_vol_max = pd.DataFrame(last_vol_group['Timepoint'].max())

last_vol_max
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_vol_max.reset_index(inplace = True)
last_vol_max['UniqueID'] = last_vol_max['Mouse ID'] + last_vol_max['Timepoint'].astype(str)
smm_nodupe['UniqueID'] = smm_nodupe['Mouse ID'] + smm_nodupe['Timepoint'].astype(str)
smm_nodupe
last_vol_merge = last_vol_max.merge(smm_nodupe, on = 'UniqueID')
last_vol_merge = last_vol_merge[['Mouse ID_x','Timepoint_x','Tumor Volume (mm3)','Metastatic Sites','Drug Regimen','Sex','Age_months','Weight (g)']]
last_vol_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
labels = last_vol_merge['Drug Regimen'].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []   
# Create 'Outlier' to store status of outlier check
last_vol_merge['Outlier'] = ""
# start loop on drugs
for drug in labels:
    # create list of tumor volumes for each drug
    tumor_vol = last_vol_merge.loc[last_vol_merge['Drug Regimen'] == drug,'Tumor Volume (mm3)']
    # calculate quartiles
    quartiles = tumor_vol.quantile([.25,.5,.75])
    # set upper and low quartiles
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    # calculate iqr
    iqr = upperq-lowerq
    # calculate upper and lower bounds for outlier check
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    # Check for outliers, store status in 'Outlier'
    for index, row in last_vol_merge.iterrows():
        if last_vol_merge.loc[index, 'Drug Regimen'] == drug and last_vol_merge.loc[index,'Tumor Volume (mm3)'] <= lower_bound:
                last_vol_merge.loc[index,'Outlier'] = True
        elif last_vol_merge.loc[index, 'Drug Regimen'] == drug and last_vol_merge.loc[index,'Tumor Volume (mm3)'] >= upper_bound:
                last_vol_merge.loc[index,'Outlier'] = True
# Create subsets
inf = last_vol_merge[last_vol_merge['Drug Regimen'] == 'Infubinol']
cef = last_vol_merge[last_vol_merge['Drug Regimen'] == 'Ceftamin']
ram = last_vol_merge[last_vol_merge['Drug Regimen'] == 'Ramicane']
cap = last_vol_merge[last_vol_merge['Drug Regimen'] == 'Capomulin']
data = [inf['Tumor Volume (mm3)'], cef['Tumor Volume (mm3)'],ram['Tumor Volume (mm3)'], cap['Tumor Volume (mm3)']]




    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
marker = dict(markerfacecolor='b', marker='*')
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume Across Selected Drug Regimens')
ax1.set_ylabel('Volume (mm3)')
ax1.boxplot(data, labels = labels,flierprops = marker)
```
